```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
```

```python
a = np.linspace(-1.3, 1.3, 100)
b = np.linspace(-1.3, 1.3, 100)
y = lambda a, b: a**4 -2*a**2 + b**2

A, B = np.meshgrid(a, b)
zs = np.array([y(u, v) for u, v in zip(np.ravel(A), np.ravel(B))])
Z = zs.reshape(A.shape)
```

```python
f = plt.figure(figsize=(15, 8))
ax = plt.axes(projection='3d')

ax.plot_surface(A, B, Z, cmap='cool', alpha=0.8)
ax.scatter3D(
    [0., 0., 1.], 
    [0., .6, .8], 
    [y(0., 0.), y(0., .6), y(1., .8)], 
    depthshade=False, 
    marker='+', 
    color='black',
    s=30,
)

ax.text(-.05, -.05, y(0., 0.) + .1, '$S$', size=15, color='#34495E')
ax.text(-.15, .65, y(0., .6) + .05, '$A$', size=15, color='#34495E')
ax.text(.95, .75, y(1., .8) + .1, '$B$', size=15, color='#34495E')

ax.set_xlabel('a')
ax.set_ylabel('b')
ax.set_zlabel('y(a, b)')

ax.view_init(30, -50)
```
